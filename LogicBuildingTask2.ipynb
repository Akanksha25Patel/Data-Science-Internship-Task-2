Problem 1: Unique Words in a Sentence


```python
# User se sentence lena
sentence = input("Enter sentence: ")

# Sentence ko words me todna
words = sentence.split()

# Duplicate hataane ke liye set banana
unique_words = set(words)

# Unique words ka count nikalna
count = len(unique_words)

print("Unique words count:", count)
print("Unique words:", unique_words)
```

Problem 2: Highest Salary from Employee Data


```python
# Employee data store karna (Dictionary)
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Sabse zyada salary wala employee nikalna
highest_salary_employee = max(employees, key=employees.get)
highest_salary_value = employees[highest_salary_employee]

print(f"Highest Salary: {highest_salary_employee} - {highest_salary_value}")
```

Problem 3: Find Maximum and Minimum Values


```python
# Input list
numbers = [45, 22, 89, 10, 66]

# Hint ke mutabik: Pehle element se initialize karein
max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    
    if num > max_val:
        max_val = num
    
    if num < min_val:
        min_val = num

print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

Problem 4: Count Products Above a Price Threshold


```python
# Input list of prices
prices = [450, 1200, 899, 1500, 300]

# Counter ko 0 se start karein
count = 0

for price in prices:
    if price > 1000:
        count = count + 1 

print(f"Products above 1000: {count}")
```

Problem 5: Calculate Attendance Percentage


```python
# Example Input
attendance = ["P", "P", "A", "P", "P"]

# Hint 1: "P" ki sankhya count karein
present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1

# Hint 2: Total number of days se divide karein
total_days = len(attendance)
percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {percentage}")
```

Problem 7: Count Character Frequency


```python
text = "pythonp" # Example text to match the output 'p': 2

# Hint: Dictionary ka use karein
frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1 
    else:
        frequency[char] = 1

print(frequency)
```

Problem 9: Check if a Key Exists in a Dictionary

```python
employee_data = {"name": "Amit", "id": 101, "dept": "IT"}

# Logic: 'in' keyword ka use karein
if "name" in employee_data:
    print("Employee exists")
else:
    print("Employee does not exist")
```
